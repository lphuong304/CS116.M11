```python
import numpy as np
import sys
import thinkdsp
import matplotlib.pyplot as pyplot
from IPython.display import Audio  
from scipy.io import wavfile
from sklearn.decomposition import PCA
```

#đọc file âm thanh

```python
#đọc file âm thanh
#hàm trả về hai giá trị là rate(khoảng thời gian lấy mẫu âm thanh) và mảng dữ liệu dưới dạng dữ liệu np.array
rate, testAudio = wavfile.read('/content/ChanChamDat.wav') 
print(rate)
start = rate * 15 #demo audio Test sau 15s đầu tiên
end = start + rate * 10 # cắt 10s đầu của audio Test
print("Số chiều của dữ liệu audio gốc: {}".format(testAudio.shape))
Audio(data=testAudio[start:end, 0], rate=rate)
```

#reshape array thành matrix kích thước H x W 

```python
getAudio = testAudio[:,0]
size = len(getAudio)
getMod = 512 - np.mod(size, 512)
getPad = np.lib.pad(getAudio, (0, getMod), 'constant', constant_values=0)

# Reshape the signal to have 1024 dimensions
reshapeAudio = getPad.reshape((len(getPad) // 512, 512))
```

#ap dụng PCA để giảm số chiều từ H xuống h (h < H) và khôi phục lại đoạn nhạc ban đầu

```python
# vận dụng kỹ thuật pca để giảm số chiều dữ liệu array audio
pca = PCA(n_components=32)
pca.fit(reshapeAudio)
transformAudio = pca.transform(reshapeAudio)
```

```python
 #khôi phục lại array audio ban đầu
reconstructAudio = pca.inverse_transform(transformAudio).reshape((len(getPad)))
```

#nghe lại đoạn Audio sau khi khôi phục

```python
print("Số chiều của dữ liệu audio sau khi khôi phục: {}".format(reconstructAudio.shape))
Audio(data=reconstructAudio[start:end], rate=rate)
```

#save file Audio

```python
with open('/content/result.wav', 'wb') as f:
    f.write(reconstructAudio[start:end].data)
```
