#**Bước 2+3+4**

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

columnTransformer = ColumnTransformer([('encoder', OneHotEncoder(), [1,2])], remainder='passthrough')

x_trans = np.array(columnTransformer.fit_transform(x_data), dtype = np.float)
```

```python
x_trans
```

```python
x_trans[5]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_trans, np.array(y_data), test_size=0.2)
```

```python
X_train
```

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler().fit(x_trans)
```

```python
X_train
```

```python
X_train.shape
```

#**Bước 5: Huấn luyện mô hình**

```python
from keras.models import Sequential
from keras.layers import Dense
def get_model():
  model = Sequential()
  model.add(Dense(6, input_shape=(13,), activation='relu'))
  model.add(Dense(6, activation='relu'))
  model.add(Dense(1, activation='sigmoid'))
  model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
  return model
```

```python
hist = model.fit(scaler.transform(X_train),y_train,epochs=100, batch_size=32)
```

#**Bước 6: Độ chính xác**

```python
model.evaluate(scaler.transform(X_test),y_test)
```

#**Bước 8: Kfold = 5**

```python
def kFoldSplit(X, y, k):
  kFold = KFold(n_splits=k)
  kFold.get_n_splits(X)

  list_X_train, list_y_train, list_X_test, list_y_test = [], [], [], []

  for i_train, i_test in kFold.split(X):
    list_X_train.append(X[i_train])
    list_y_train.append(y[i_train])
    list_X_test.append(X[i_test])
    list_y_test.append(y[i_test])

  return list_X_train, list_y_train, list_X_test, list_y_test
```

```python
list_X_train, list_y_train, list_X_test, list_y_test  = kFoldSplit(scaler.transform(x_trans), np.array(y_data), 5)
```

```python
def kfold_rslt(model):
  list_accuracy = []
  for i in range(len(list_X_train)):
    hist = model.fit(list_X_train[i], list_y_train[i],epochs=100,batch_size=32)
    list_accuracy.append(hist.history['accuracy'][-1])
  return list_accuracy
```

```python
list_acc_score = kfold_rslt(get_model())
```
