# **Bài làm**

### **Các hàm và thư viện cần thiết**

```python
import cv2 as cv
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
import numpy as np
from sklearn.decomposition import PCA
```

## **1. Đọc ảnh mức xám từ file ảnh kích thước (h, w)**

```python
# Tải ảnh bằng GDown
!gdown --id "1jknCnh7fYb5GBos-1ka8eDo1IGfwWiyG"

# Gán path ảnh
path_image = "/content/lena.jpg"

# Đọc ảnh mức xám
image = cv.imread(path_image, 0)

# Đọc kích thước ảnh
h, w = image.shape
```

## **2. Hiển thị ảnh**

```python
# Hiển thị ảnh
cv2_imshow(image)
```

## **3. Dùng PCA để giảm chiều từ H -> h chiều**

```python
X = np.array(image)
k = 64
pca = PCA(n_components=k)
pca.fit(X)
imagePCA = pca.transform(image).T

print("Ảnh gốc: ", image.shape)
print("Ảnh sau PCA:" , imagePCA.shape)
```

## **4. Khôi phục lại ảnh gốc**

```python
imageRecovery = pca.inverse_transform(imagePCA.T)
```

```python
print("Ảnh gốc: ", image.shape)
print("Ảnh khôi phục:" , imageRecovery.shape)
```

## **5. Hiển thị lại ảnh đã khôi khục**

```python
cv2_imshow(np.hstack((image, imageRecovery)))
```
