#import thư viện cần thiết

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA
from google.colab.patches import cv2_imshow
from sklearn.feature_extraction import image
```

```python
img = cv2.imread('/content/Bicycle.jpg',0)
```

#vận dụng thuật toán PCA cho ảnh

```python
pca = PCA(n_components=100)
pca.fit(img)

imgPCAbefore = np.dot(img, pca.components_.T) # giảm chiều dữ liệu ảnh
imgPCAafter = np.dot(imgPCAbefore, pca.components_) # khôi phục dữ liệu ảnh
```

```python
print('Số chiều dữ liệu ảnh ban đầu: {}'.format(img.shape))
print('Số chiều dữ liệu ảnh sau khi vận dụng PCA: {}'.format(imgPCAbefore.shape))
print('Số chiều dữ liệu ảnh từ ảnh được PCA về chiều ảnh dữ liệu ban đầu: {}'.format(imgPCAafter.shape))
```

#đọc hiển thị 2 ảnh

```python
cv2_imshow(img) #hiển thị ảnh gốc
cv2_imshow(imgPCAbefore.T) #hiển thị ảnh đã giảm chiều dữ liệu
cv2_imshow(imgPCAafter) #hiển thị ảnh khôi phục sau khi đã giảm chiều dữ liệu 
```
